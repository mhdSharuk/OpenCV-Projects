```
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```
img = cv2.imread('/content/cam.jpg')
plt.xticks([])
plt.yticks([])
plt.imshow(img)
```

# Smoothing

**2D Conv Filter**

```
kernel = np.ones((3,3),np.float64)
kernel
```

```
conv_filter = cv2.filter2D(img,-1,kernel)
```

```
plt.figure(figsize = (10,9))
plt.subplot(1,2,1)
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Conv Filter')
plt.xticks([])
plt.yticks([])
plt.imshow(conv_filter)
```

**Averaging**

```
blur = cv2.blur(img,(3,3))
```

```
plt.figure(figsize = (10,9))
plt.subplot(1,2,1)
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Blur')
plt.xticks([])
plt.yticks([])
plt.imshow(blur)
```

**Gausssian Filtering**

```
gaussian = cv2.GaussianBlur(img,(3,3),3)
```

```
plt.figure(figsize = (10,9))
plt.subplot(1,2,1)
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Gaussian Blur')
plt.xticks([])
plt.yticks([])
plt.imshow(gaussian)
```

**Median Blur**

```
med_blur = cv2.medianBlur(img,5)
```

```
plt.figure(figsize = (15,10))
plt.subplot(1,2,1)
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('Median Blur')
plt.xticks([])
plt.yticks([])
plt.imshow(med_blur)
```

**CLAHE**

```
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl = clahe.apply(img_gray)
```

```
plt.figure(figsize = (15,10))
plt.subplot(1,2,1)
plt.title('Original')
plt.xticks([])
plt.yticks([])
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('CLAHE')
plt.xticks([])
plt.yticks([])
plt.imshow(cl)
```
